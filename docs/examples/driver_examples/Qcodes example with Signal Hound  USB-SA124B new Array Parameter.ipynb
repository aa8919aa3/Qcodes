```python
import qcodes as qc
```

```python
from qcodes.instrument_drivers.signal_hound.USB_SA124B import SignalHound_USB_SA124B
from qcodes.dataset.measurements import Measurement
from qcodes.dataset.plotting import plot_by_id
```

```python
mysa = SignalHound_USB_SA124B('mysa', dll_path='C:\\Program Files\\Signal Hound\\Spike\\sa_api.dll')
```

```python
mysa.get_idn()
```

```python
mysa.freq_sweep()
```

```python
mysa.frequency(2e9)
mysa.span(0.5e6)
```

```python
mysa.avg(1)
meas = Measurement()
meas.register_parameter(mysa.freq_sweep)  # register the first independent parameter
with meas.run() as datasaver:
    datasaver.add_result((mysa.freq_sweep, mysa.freq_sweep.get(),))
    
    dataid = datasaver.run_id  # convenient to have for plotting
plot_by_id(dataid)
```
